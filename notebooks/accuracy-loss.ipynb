```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Upload from CSV
training_metrics = '../data/2024-09-22-conv3d-iter3.csv'
training_df = pd.read_csv(training_metrics)
training_df
```

```python
# Plot 
accuracy = training_df['accuracy']
loss = training_df['loss']
val_accuracy = loss = training_df['val_accuracy']
val_loss = training_df['val_loss']

fig, axs = plt.subplots(1, 2, figsize=(8, 3.5), layout='tight')

# Set axs
axs[0].plot(training_df['epoch'], training_df['accuracy'], label = f"accuracy")
axs[0].plot(training_df['epoch'], training_df['loss'], label = f"loss")
axs[1].plot(training_df['epoch'], training_df['val_accuracy'], label = f"val_accuracy")
axs[1].plot(training_df['epoch'], training_df['val_loss'], label = f"val_loss")

# Set titles
plt.suptitle("Conv3D: Accuracy and Loss", fontsize=14)
axs[0].set_title("accuracy / loss", fontsize=11)
axs[1].set_title("val_accuracy / val_loss", fontsize=11)

# Set labels
axs[1].set_ylabel('Accuracy / Loss')
axs[1].set_ylabel('Accuracy / Loss')
axs[0].set_xlabel('Epoch')
axs[1].set_xlabel('Epoch')

plt.legend()

# Show plot
plt.show()
```
