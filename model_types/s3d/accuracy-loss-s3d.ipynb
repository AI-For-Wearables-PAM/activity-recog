# Plot training metrics for S3D

````python
import pandas as pd
import matplotlib.pyplot as plt
````

````python
# Upload from CSV
training_metrics = '../data/2024-10-03-pretrained-acc-loss.csv'
training_df = pd.read_csv(training_metrics)
training_df
````

## Create plot

````python
# Plot 
accuracy = training_df['accuracy']
loss = training_df['loss']

fig, axs = plt.subplots(2, 1, figsize=(4, 6), layout='tight')

# Set axs
axs[0].plot(training_df['epoch'], training_df['accuracy'], label = f"accuracy", color='blue')
axs[0].legend()
axs[1].plot(training_df['epoch'], training_df['loss'], label = f"loss", color='orange')

# Set title
plt.suptitle("S3D: Accuracy and Loss", fontsize=14)

# Set labels
axs[0].set_ylabel('Accuracy / Loss')
axs[0].set_ylabel('Accuracy')
axs[1].set_ylabel('Loss')
axs[1].set_xlabel('Epoch')

axs[1].legend()

# Show plot
plt.show()
````
